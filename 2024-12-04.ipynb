```python
import re 
```

```python
total = 0
```

```python
with open('input-03.txt', 'r') as file: 
    text = file.read() 
```

```python
pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
matches = re.findall(pattern, text)
    
for match in matches: 
    print(match)
    x = int(match[0])
    y = int(match[1])
    result = x * y 
    total += result
        
print(total)
```

```python
total = 0
enabled = True
    
mul_pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
do_pattern = r'do\(\)' 
dont_pattern = r'don\'t\(\)'

instructions = []
    
for match in re.finditer(mul_pattern, text): 
    x, y = int(match.group(1)), int(match.group(2))
    instructions.append((match.start(), 'mul', x * y))
        
for match in re.finditer(do_pattern, text): 
    instructions.append((match.start(), 'do', None))
    
for match in re.finditer(dont_pattern, text): 
    instructions.append((match.start(), 'dont', None)) 
    
instructions.sort(key=lambda x: x[0])
    
for pos, type_name, value in instructions: 
    if type_name == 'do': 
        enabled = True 
    elif type_name == 'dont': 
        enabled = False 
    elif type_name == 'mul' and enabled: 
        total += value 
            
print(total)
```
